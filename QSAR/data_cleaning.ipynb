```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('clean.csv')

def act(num):
    if num > 100:
        return 0
    else:
        return 1
    
df['Activity'] = [act(value) for value in df['Standard Value'].values.tolist()]
df.to_csv('Class.csv')
```
