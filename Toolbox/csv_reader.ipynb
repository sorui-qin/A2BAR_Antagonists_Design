```python
import pandas as pd
```

```python
csv_file = r'D:\Research\A2B\Meastro\MMPBSA-Best.csv'
df = pd.read_csv(csv_file)
df
```

```python
tolist = lambda index: df[index].values.tolist()
docking_score = list(zip(tolist('Title'), tolist('docking score')))
docking_sorted = sorted(docking_score, key= lambda i : i[-1])

mmgbsa = list(zip(tolist('Title'), tolist('MMGBSA dG Bind')))
mmgbsa_sorted = sorted(mmgbsa, key= lambda i : i[-1])

dock_idx = [line[0] for line in docking_sorted][:30]
mmgbsa_idx = [line[0] for line in mmgbsa_sorted][:30]

combined_idx = [idx for idx in dock_idx if idx in mmgbsa_idx]
```

```python
combined_idx
```

```python
with open(r'D:\Research\A2B\index.txt', 'w+') as wr:
    for i in combined_idx:
        wr.write(i+'\n')
    wr.close()
```
