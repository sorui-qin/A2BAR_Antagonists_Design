```python
from scipy.interpolate import make_interp_spline
import numpy as np
from matplotlib import pyplot as plt
```

```python
filename = r'D:\Research\A2B\training\loss_value.txt'
with open(filename, 'r+') as fi:
    value_li = fi.read().split('\n')
    fi.close()
value_arr = np.array(value_li).astype(np.float64)
value_arr
```

```python
minus_li = [abs(eval(value_li[i+1]) - eval(value_li[i])) for i in range(len(value_li)-1)][1:]
D_value_arr = np.array(minus_li)
```

```python
x1_axis = np.array([num for num in range(1,len(value_li)+1)])
loss_axis = value_arr
plt.figure(figsize=(6,5),dpi=300)
plt.plot(x1_axis,loss_axis,label='Loss curve')

plt.xticks(range(0,100,10)) #Axis justify
plt.yticks([2,2.7,4,6,8])
plt.xlim((0, 95))
plt.xlabel("Epoches",fontsize=13)
plt.ylabel("Loss value",fontsize=13)
plt.title("Loss value",fontsize=15)

threshold = 2.7 
plt.axhline(threshold, color='red', lw=1, alpha=0.7, ls='--',label='Convergence')
cutoff = 60
plt.axvline(cutoff, color='green', lw=1, alpha=0.7, ls='dashdot')

plt.legend(loc='best')
plt.show
```

```python
x2_axis = np.array([num for num in range(1,len(value_li)-1)])
plt.figure(figsize=(6,6),dpi=300)
plt.plot(x2_axis,D_value_arr,label='D-vaule')

plt.yscale('log')
plt.xticks(range(0,100,10))
plt.xlim((0, 95))

threshold = 1e-2
plt.axhline(threshold, color='red', lw=1, alpha=0.7, ls='--',label='Thershold')
cutoff = 60
plt.axvline(cutoff, color='green', lw=1, alpha=0.7, ls='dashdot')

plt.xlabel("Epoches",fontsize=13)
plt.ylabel("D-value",fontsize=13)
plt.title("D-value of loss value between each epoches",fontsize=15)

plt.legend(loc='best')
plt.show
```
