```python
filename = r'D:\Research\A2B\Molecules\mmgbsa_screened_97.smi'
with open(filename, 'r+') as fi:
    origin_li = fi.read().split('\n')
split_li = [lines.split(' ') for lines in origin_li][:-1]
split_li
```

```python
#Directly take the smiles from the file
smi_li = [lines[0] for lines in split_li]
filename_processed = filename.split('.')[0] + '_smiles.smi'
with open(filename_processed, 'w+') as wr:
    for smi in smi_li:
        wr.write(smi+'\n')
    wr.close()
```

```python
#Find specific smiles by given index
smi_di = {}
for line in split_li:
    smi_di[line[1]] = line[0]

index = r'D:\Research\A2B\index.txt'
with open(index, 'r') as id:
    index_li = id.read().split('\n')

found_li = []
for ids in index_li:
    found_li.append(smi_di[ids])

filename_processed = filename.split('.')[0] + '_index.smi'
with open(filename_processed, 'w+') as wr:
    for smi in found_li:
        wr.write(smi+'\n')
    wr.close()
```
