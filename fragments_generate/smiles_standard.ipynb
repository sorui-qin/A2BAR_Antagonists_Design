```python
import pandas as pd 
import rdkit
import rdkit.Chem as rkc
```

```python
def standard_smi(smile):
    return rkc.MolToSmiles(rkc.MolFromSmiles(smile))

def only_one_dummyatom(smile):
    dummynum = smile.count('*')
    if dummynum != 1:
        return None
    else:
        return smile
```

```python
df=pd.read_csv(r'D:\科研\微科研\片段库\ChEMBL_Fragments.csv',usecols=[2])
frag=df.drop_duplicates().values.tolist()
```

```python
frag_list=[]
for i in frag:
    for smi in i:
        if only_one_dummyatom(smi):
            standard_smi(smi)
            frag_list.append(smi) 
frag_list   
```

```python
sta_smi=[standard_smi(smi) for smi in frag_list]
with open('ChemBL_standard.txt',"a+") as f:
    for i in sta_smi:
        f.write(i+'\n')
```
