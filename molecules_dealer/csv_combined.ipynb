```python
import pandas as pd
import rdkit
import rdkit.Chem as rkc

filename = r'D:\Research\A2B\Molecules\benzyl_meta\gene_benzyl_meta.csv'

def standrad_smi(smile):
    return rkc.MolToSmiles(rkc.MolFromSmiles(smile))

def valid_smi(smi):
    if  "*" in smi or smi == 'smiles':
        pass
    else:
        return smi

def delete_same(li):
    fin_li=[]
    for smi in li:
        if valid_smi(smi):
            fin_li.append(valid_smi(standrad_smi(smi)))
    del_li = sorted(set(fin_li),key=fin_li.index)
    return del_li
```

```python
df = pd.read_csv(filename)
df
```

```python
standrad_li = delete_same(df["smiles"].values.tolist())
standrad_li
```

```python
with open(filename.replace('csv','smi'),'w+') as wr:
    for smi in standrad_li:
        wr.write(smi+'\n')
    wr.close
```
