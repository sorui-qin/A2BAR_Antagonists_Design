```python
import rdkit.Chem as rkc
from rdkit.Chem import rdRGroupDecomposition as rdRGD
import csv
import pandas as pd
```

```python
filename = r'D:\Research\A2B\Molecules\benzyl_meta\benzylmeta-gene_R1filted.smi'
with open(filename,'r+') as load_smi:
    smi_li = load_smi.read().split('\n')
mols = [rkc.MolFromSmiles(smi) for smi in smi_li]
```

```python
sca =rkc.MolFromSmiles('NC1=NC(CC2=CC=CC=C2)=CN3C4=CC=CC=C4N=C31')
res,unmatched = rdRGD.RGroupDecompose([sca],mols,asSmiles=True)
res
```

```python
len(smi_li)
```

```python
for num, smi in enumerate(smi_li):
    try:
        res[num]['smiles'] = smi
    except IndexError:
        pass
len(res)
```

```python
def substitutent_filter(smi): #Count the non-ring carbon number
    if smi.count('C') < 4:
        if 'c' in smi or 'N' in smi  or 'S' in smi  or 'O-'in smi:
            pass
        else:
            query = rkc.MolFromSmarts('O')
            if not rkc.MolFromSmiles(smi).HasSubstructMatch(query):
                return smi

filted_smi = []
for sub in res:
    if substitutent_filter(sub['R1']):
        filted_smi.append(sub['smiles'])
len(filted_smi)
```

```python
with open(r'D:\Research\A2B\Molecules\benzyl_meta\benzylmeta-gene_R1filted.smi','w+') as fi:
    for smi in filted_smi:
        fi.write(smi+'\n')
    fi.close()
```
